```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting

plt.figure(figsize = (10,6))
common_category = df['product_tag'].value_counts().head(10)
sns.barplot(x = common_category.values, y = common_category.index, palette = 'Blues_d')
plt.title('Top 10 Most Common Product Categories')
plt.xlabel('Number of Products')
plt.ylabel('Product Category')
plt.show()
```

## Q3) Which brand provides the highest average discounts

```python
top_10_brands = df['brand_tag'].value_counts().head(10).index
df[df['brand_tag'].isin(top_10_brands)].groupby('brand_tag')['discount_percent'].mean().sort_values(ascending = False)
```

## Q4) Is there any relationship between product price and customer ratings

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = 'marked_price', y = 'rating', hue = 'discount_percent', data = df, palette = 'coolwarm', alpha = 0.6)
plt.title('Price vs Rating with Discount Percentage')
plt.xlabel('Marked Price')
plt.ylabel('Rating')
plt.show()
```

Key Insights:

* No Clear Price-Rating Relationship: In many cases, you might observe that the price of a product does not necessarily correlate with its rating. For instance, expensive products might not always have higher ratings compared to cheaper ones, indicating that customer satisfaction isn't purely tied to price.
* Price Outliers:
You may find products with extreme prices (either very high or very low) having diverse ratings. This could suggest that price alone is not a significant factor in determining customer satisfaction or perceived product quality.
* Discounted Products and Ratings:
By incorporating the discount_percent as the hue in the plot, you could analyze whether discounted products tend to have higher or lower ratings. If discounted products cluster around lower ratings, it might indicate that discounts are applied to clear less desirable stock.
* Popular Price Ranges:
Some price ranges may have a higher density of ratings. For example, mid-priced products (1000-5000) might gather more reviews and ratings, indicating higher popularity or broader affordability among customers.
* Price Threshold for High Ratings:
You might identify a price threshold beyond which ratings plateau or drop. For instance, after a certain price (e.g., 5000), ratings may not increase, indicating diminishing returns on customer satisfaction for very high-priced items.
* Low Ratings for Expensive Products:
If you observe clusters of high-priced products with low ratings, it could point to customer dissatisfaction, possibly due to unmet expectations in product quality or value for money.
